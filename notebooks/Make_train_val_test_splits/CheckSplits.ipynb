```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt

from datasets.data_retriever_and_generator import get_img_and_gt
```

```python
dataset_path = 'D:\\ThesisData\\Datasets\\ShanghaiTech\\Part_B'
data_path = os.path.join(dataset_path, 'test_split.csv')
gen_den_key = 'SHT'
```

```python
data = pd.read_csv(data_path)
data = data.to_numpy()
```

```python
for rel_img_path, rel_gt_path in data:
    img_path = os.path.join(dataset_path, rel_img_path)
    gt_path = os.path.join(dataset_path, rel_gt_path)
    img, gt = get_img_and_gt(img_path, gt_path, gen_den_key)
    
    img = np.array(img)
    gt = np.array(gt)
    
    gt_mask = gt.copy()
    gt_mask[gt_mask > 0] = 1
    gt_mask = gt_mask.astype(np.uint8)
    img[:, :, 0][gt_mask == 1] = 255
    img[:, :, 1][gt_mask == 1] = 255
    img[:, :, 2][gt_mask == 1] = 255

    print((rel_img_path, rel_gt_path))
    
    plt.figure(figsize=(15, 15))
    plt.imshow(img)
    plt.show()
    
    
    input()
    
    
```
